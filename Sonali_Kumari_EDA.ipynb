### Data Visualization

```python
#Customer Purchase Frequency by Region
region_purchase_count = final_df.groupby('Region')['TransactionID'].nunique().reset_index()
sns.barplot(x='Region', y='TransactionID', data=region_purchase_count)
plt.title('Number of Transactions by Region')
plt.xlabel('Region')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.show()
```

```python
#Total Sales Value by Product Category
category_sales = final_df.groupby('Category')['TotalValue'].sum().reset_index()
sns.barplot(x='Category', y='TotalValue', data=category_sales)
plt.title('Total Sales Value by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales Value')
plt.xticks(rotation=45)
plt.show()
```

```python
product_quantity = final_df.groupby('ProductName')['Quantity'].sum().reset_index().sort_values( by = 'Quantity', ascending = False)
product_quantity.head(5)
```

```python
#Top-Selling Products by Total Sales Value
top_selling_products = final_df.groupby('ProductName')['TotalValue'].sum().reset_index().sort_values(by='TotalValue', ascending=False)
sns.barplot(x='TotalValue', y='ProductName', data=top_selling_products.head(10), orient='h')
plt.title('Top-Selling Products by Total Sales Value')
plt.xlabel('Total Sales Value')
plt.ylabel('Product Name')
plt.show()
```

```python
final_df['TotalValue'] = pd.to_numeric(final_df['TotalValue'], errors='coerce')
final_df['TransactionDate'] = pd.to_datetime(final_df['TransactionDate'])
final_df['Month'] = final_df['TransactionDate'].dt.to_period('M').dt.to_timestamp()

# Aggregate total sales per month
monthly_sales = final_df.groupby('Month')['TotalValue'].sum().reset_index()

# Aggregate sales by region per month
region_sales_over_time = final_df.groupby(['Month', 'Region'])['TotalValue'].sum().reset_index()

# Plot the total sales trend over time
sns.lineplot(x='Month', y='TotalValue', color='y', data=monthly_sales, label='Total Sales')

# Plot sales trend by region over time
sns.lineplot(x='Month', y='TotalValue', hue='Region', data=region_sales_over_time)

# Add title and labels
plt.title('Sales Trend Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales Value')
plt.legend(title='Region', loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
#region wise category sales value
region_category_sales = final_df.groupby(['Region', 'Category'])['TotalValue'].sum().reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='Region', y='TotalValue', hue='Category', data=region_category_sales, palette='Set2')
plt.title('Sales of Product Categories in Different Regions')
plt.xlabel('Region')
plt.ylabel('Total Sales Value')
plt.legend(title='Product Category', loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

```python
# signups per month for each region
final_df['SignupDate'] = pd.to_datetime(final_df['SignupDate'])
final_df['Month'] = final_df['SignupDate'].dt.to_period('M').dt.to_timestamp()
region_signup_over_time = final_df.groupby(['Month', 'Region'])['CustomerID'].count().reset_index()
total_signup_over_time = final_df.groupby('Month')['CustomerID'].count().reset_index()


plt.figure(figsize=(12, 8))
sns.lineplot(x='Month', y='CustomerID', hue='Region', data=region_signup_over_time, marker='o')
sns.lineplot(x='Month', y='CustomerID', data=total_signup_over_time, label='Total Signups', color='k', linestyle='--', marker='x')
plt.title('Signups Over Time by Region and Total')
plt.xlabel('Month')
plt.ylabel('Number of Signups')
plt.xticks(rotation=45)
plt.legend(title='Region', loc='upper left')
plt.show()
```

```python
# Group by product category and the average price for each category
category_avg_price = final_df.groupby('Category')['Price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Price', data=category_avg_price, palette='Blues_d')

plt.title('Average Price per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count customer signups per month
customer_acquisition = final_df.groupby('Month')['CustomerID'].nunique().reset_index()

plt.figure(figsize=(12, 8))
sns.lineplot(x='Month', y='CustomerID', data=customer_acquisition, color='blue')
plt.title('Customer Acquisition Over Time')
plt.xlabel('Month')
plt.ylabel('Number of New Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Salkes conversion rate (number of transactions per signup)
sales_conversion = final_df.groupby('Month').agg(
    total_signups=('CustomerID', 'nunique'),
    total_transactions=('TransactionID', 'nunique')
).reset_index()

sales_conversion['conversion_rate'] = sales_conversion['total_transactions'] / sales_conversion['total_signups']
plt.figure(figsize=(12, 8))
sns.lineplot(x='Month', y='conversion_rate', data=sales_conversion, marker='o', color='green')
plt.title('Sales Conversion Rate Over Time')
plt.xlabel('Month')
plt.ylabel('Conversion Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# Product category sales share
category_sales_share = final_df.groupby('Category')['TotalValue'].sum().reset_index()
plt.pie(category_sales_share['TotalValue'], labels=category_sales_share['Category'], autopct='%1.1f%%')
plt.title('Sales Share by Product Category')
plt.show()
```

```python
# average order value per month
final_df['OrderValue'] = final_df['Quantity'] * final_df['Price']
aov = final_df.groupby('Month')['OrderValue'].mean().reset_index()
plt.figure(figsize=(12, 8))
sns.lineplot(x='Month', y='OrderValue', data=aov, marker='o', color='purple')
plt.title('Average Order Value (AOV) Over Time')
plt.xlabel('Month')
plt.ylabel('Average Order Value')
plt.xticks(rotation=45)
plt.show()

```
